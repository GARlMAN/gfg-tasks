```python
import math


#to print each alphabet
def A(height):                                       #draws A
    A_widht = height
    n = A_widht
    for i in range(0, height):
        A_widht = 2*height
        for j in range(0, A_widht+1):
            if (j == n or j == (A_widht - n) or (i == (height // 2) and j > n and j < (A_widht - n))):
                print("*", end="")
            else:
                print(end=" ")
        print()
        n = n-1 


def B(height) :                                       #draws B
    mid = height // 2
    width = math.floor(1.5 * height)
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ((i == 0 or i == height - 1 or i == mid) and j < (width - 2)) :
                print("*",end="")
            elif (j == (width - 2) and not(i == 0 or i == height - 1 or i == mid)) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def C(height):                                       #draws C
    width = math.floor(1.5 * height) 
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if (i == 0 or i == height - 1 ) :
                print("*",end="")
            else :
                continue
        print()

def D(height):                                       #draws D
    width = math.floor(1.5 * height)
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (i == 0 or i == height - 1) and j < width - 1 ):
                print("*",end="")
            elif (j == width - 1 and i != 0 and i != width - 1) :
                print("*",end="")
            else :
                print(end=" ")
        print()


def E(height):                                       #draws E
    width = math.floor(1.5 * height)
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (i == 0 or i == height - 1 ) or (i == height // 2 and j <= width // 2) ):
                print("*",end="")
            else :
                continue
        print()


def F(height):
    width = math.floor(1.5 * height)                                       #draws F
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (i == 0) or (i == height // 2 and j <= width // 2) ):
                print("*",end="")
            else :
                continue
        print()


def G(height):                                       #draws G
    width = math.floor(1.5 * height) 
    for i in range(0,height) :
        for j in range(0,width-1) :
            if ((i == 0 or i == height - 1) and (j == 0 or j == width - 2)) :
                print(end=" ")
            elif (j == 0) :
                print("*",end="")
            elif (i == 0 and j <= height) :
                print("*",end="")
            elif (i == height // 2 and j > height // 2) :
                print("*",end="")
            elif (i > height // 2 and j == width - 2) :
                print("*",end="")
            elif (i == height - 1 and j < width - 1) :
                print("*",end="")
            else :
                print(end=" ")
        print()


def H(height) :                                       #draws H
    width = math.floor(1.5 * height)     
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (j == width - 1) or (i == height // 2) ):
                print("*",end="")
            else :
                print(end=" ")
        print()

def I(height) :                                       #draws I
    width = math.floor(1.5 * height)     
    for i in range(0,height) :
        for j in range(0,width) :
            if ( i == 0 or i == height - 1 ):
                print("*",end="")
            elif ( j == width // 2 ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def J(height) :                                       #draws J
    width = math.floor(1.5 * height)     
    for i in range(0,height) :
        for j in range(0,width) :
            if ( i == height - 1 and (j > 0 and j < width - 1) ):
                print("*",end="")
            elif ( (j == width - 1 and i != height - 1) or (i > (height // 2) - 1 and j == 0 and i != height - 1) ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def K(height) :                                       #draws K
    half = height // 2
    dummy = half
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,half+1) :
            if ( j == abs(dummy) ):
                print("*",end="")
            else :
                print(end=" ")
        print()
        dummy = dummy -1

def L(height) :                                       #draws L
    width = math.floor(1.5 * height)     
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( i == height - 1 ):
                print("*",end="")
            else :
                print(end=" ")
        print()

def M(height) :                                       #draws M
    counter = 0
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,height+1) :
            if ( j == height ):
                print("*",end="")
            elif ( j == counter or j == height - counter - 1 ) :
                print("*",end="")
            else :
                print(end=" ")
        if(counter == height // 2) :
            counter = -99999
        else :
            counter = counter + 1
         
        print()

def N(height) :                                       #draws N
    counter = 0
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,height+1) :
            if ( j == height ):
                print("*",end="")
            elif ( j == counter) :
                print("*",end="")
            else :
                print(end=" ")
        counter = counter + 1
        print()

def O(height):                                       #draws O
    width = math.floor(1.5 * height)
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (i == 0 or i == height - 1) and j < width - 1 ):
                print("*",end="")
            elif (j == width - 1 and i != 0 and i != width - 1) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def P(height) :                                       #draws P
    width = math.floor(1.5 * height) 
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (i == 0 or i == height // 2) and j < width - 1 ):
                print("*",end="")
            elif ( i < height // 2 and j == width - 1 and i != 0 ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def Q(height):                                       #draws Q
    O(height)

def R(height) :                                       #draws R
    width = math.floor(1.5 * height) 
    half = (height // 2)
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,width) :
            if ( (i == 0 or i == half) and j < (width - 2) ):
                print("*",end="")
            elif ( j == (width - 2) and not(i == 0 or i == half) ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def S(height) :                                       #draws S
    width = math.floor(1.5 * height) 
    for i in range(0,height) :
        for j in range(0,width) :
            if ( (i == 0 or i == height // 2 or i == height - 1) ):
                print("*",end="")
            elif ( i < height // 2 and j == 0 ) :
                print("*",end="")
            elif ( i > height // 2 and j == width - 1 ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def T(height) :                                       #draws T
    width = math.floor(1.5 * height) 
    for i in range(0,height) :
        for j in range(0,width) :
            if ( i == 0 ):
                print("*",end="")
            elif ( j == width // 2 ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def U(height) :                                       #draws U
    width = math.floor(1.5 * height) 
    for i in range(0,height) :
        if (i != 0 and i != height - 1) :
            print("*",end="")
        else :
            print(end = " ")
        for j in range(0,width) :
            if ( ((i == height - 1) and j >= 0 and j < width - 1) ):
                print("*",end="")
            elif ( j == width - 1 and i != 0 and i != height - 1 ) :
                print("*",end="")
            else :
                print(end=" ")
        print()

def V(height) :                                       #draws V
    counter = 0
    width = height*2
    for i in range(0,height) :
        for j in range(0,width+1) :
            if ( j == counter or j == width - counter - 1 ):
                print("*",end="")
            else :
                print(end=" ")
 
        counter = counter + 1
        print()

def W(height) :                                       #draws W
    counter = height // 2
    for i in range(0,height) :
        print("*",end="")
        for j in range(0,height+1) :
            if ( j == height ):
                print("*",end="")
            elif ( (i >= height // 2) and (j == counter or j == height - counter - 1) ) :
                print("*",end="")
            else :
                print(end=" ")
        if( i >= height // 2) :
            counter = counter + 1
        print()


def X(height) :                                       #draws X
    counter = 0
    for i in range(0,height+1) :
        for j in range(0,height+1) :
            if ( j == counter or j == height - counter ):
                print("*",end="")
            else :
                print(end=" ")
        counter = counter + 1
        print()

def Y(height):                                       #draws Y
    counter = 0
    for i in range(0,height) :
        for j in range(0,height+1) :
            if ( j == counter or j == height - counter and i <= height // 2 ):
                print("*",end="")
            else :
                print(end=" ")
        print()
        if (i < height // 2) :
            counter = counter + 1

def Z(height) :                                       #draws Z
    counter = height - 1
    for i in range(0,height) :
        for j in range(0,height) :
            if ( i == 0 or i == height - 1 or j == counter ):
                print("*",end="")
            else :
                print(end=" ")
        counter = counter - 1
        print()
```
