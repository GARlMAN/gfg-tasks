```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.isnull().sum()
data.drop("date", axis=1, inplace=True)
data.drop("is_weekend", axis=1, inplace=True)
data.drop("hour", axis=1, inplace=True)
```

```python
data.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop("number_people", axis=1),data['number_people']  , test_size=0.8, random_state=42)
```

```python
X_test
```

```python
y_train
```

```python
normalization = StandardScaler()
```

```python
normalization.fit(X_train)
```

```python
X_train = normalization.transform(X_train)
X_test = normalization.transform(X_test)
```

```python
X_train
```

```python
epocs = []
C_value = [100, 10, 1, 0.1, 0.01, 0.001]
score_on_v_of_C = []
```

```python
SVR_model = SVR(C=1, cache_size=2500, epsilon=2, kernel='rbf')
```

```python
for i in C_value:
    SVR_model.set_params(C=i)
    SVR_model.fit(X_train, y_train)
    score_on_v_of_C.append(SVR_model.score(X_test, y_test))
```
